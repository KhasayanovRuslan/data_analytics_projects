- `churn` — ушёл или нет.

### 2.3. Рассмотреть типы данных в каждом столбце

### 2.4. Проверить наличие дубликатов

```python
bank_scrooge.duplicated().sum()
```

Имеется 1 полный дубликат.

```python
# Удаление дубликата
bank_scrooge = bank_scrooge.drop_duplicates()
```

### 2.5. Кодирование данных

```python
# Создаем копии столбцов city и gender
bank_scrooge['city_cp'] = bank_scrooge['city'].copy()
bank_scrooge['gender_cp'] = bank_scrooge['gender'].copy()
bank_scrooge.head()
```

```python
# Осуществляем кодирование данных полей gender и city
bank_scrooge = pd.get_dummies(bank_scrooge, prefix=['city_', 'gender_'], columns=['city_cp', 'gender_cp'])
bank_scrooge.head()
```

## 3. Исследовательский анализ данных

### 3.1. Построить распределения по разным полям, в т. ч. в разрезе оттока

```python
# Определим, как распределяются клиенты по городам
clients_city = bank_scrooge.groupby('city', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'})

display(clients_city)

sns.barplot(x='city', y='clients_count', data=clients_city)

plt.title('Количество клиентов по городам')
plt.xlabel('Город')
plt.ylabel('Количество клиентов')

plt.show()
```

Наибольшее количество клиентов банка находится в Ярославле(5864), наименьшее - в Ростове(1421).

```python
# Определим, как распределяются клиенты по городам и полу
clients_city_gender = bank_scrooge.groupby(by=['city', 'gender'], as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'})

display(clients_city_gender)

sns.barplot(x='city', y='clients_count', hue='gender', data=clients_city_gender)

plt.title('Количество клиентов по городам и полу')
plt.xlabel('Город')
plt.ylabel('Количество клиентов')

plt.show()
```

Соотношение клиентов по полу в разных городах почти одинаковое, есть только незначительные различия: в Ростове и Рыбинске среди клиентов преобладают женщины, в Ярославле - мужчины.

```python
# Вычислим процент отточных клиентов по городам
city_churn = bank_scrooge.groupby('city', as_index=False).agg({'churn': 'sum', 'user_id':'count'})\
.rename(columns={'churn': 'churn_count', 'user_id': 'clients_count'})

city_churn['churn_percentage'] = (city_churn['churn_count'] * 100 / city_churn['clients_count']).round(2)
city_churn = city_churn.sort_values(by='churn_percentage', ascending=False)

display(city_churn)

sns.barplot(x='city', y='churn_percentage', data=city_churn)

plt.title('процент отточных клиентов по городам')
plt.xlabel('город')
plt.ylabel('процент')
plt.show()
```

Процент отточных клиентов по каждому городу находится в диапазоне от 16% до 19%. Наибольшее количество таких клиентов в Ярославле, наименьшее - в Рыбинске. В Ростове чуть меньше отточных клиентов, чем в Ярославле; в обоих городах приблизительно 19% таких клиентов.

```python
# Вычислим процент отточных клиентов по городам и полу
city_gender_churn = bank_scrooge.groupby(by=['city', 'gender'], as_index=False)\
.agg({'churn': 'sum', 'user_id':'count'}).rename(columns={'churn': 'churn_count', 'user_id': 'clients_count'})

city_gender_churn['churn_percentage'] = (city_gender_churn['churn_count'] * 100 / city_gender_churn['clients_count']).round(2)

display(city_gender_churn)

sns.barplot(x='city', y='churn_percentage', hue='gender', data=city_gender_churn)

plt.title('процент отточных клиентов по городам и полу')
plt.xlabel('город')
plt.ylabel('процент')
plt.show()
```

По всем городам среди отточных клиентов больше мужчин в 1,5-2 раза.

На данный момент мы выяснили, что наибольшее количество отточных клиентов наблюдается в Ярославле и Ростове среди мужчин. Посмотрим, как распределяется возраст для данных клиентов, а также как распределяется этот признак для неотточных клиентов. 

```python
bank_scrooge.query('city == "Ярославль" and gender == "М" and churn == 1')['age'].hist()

plt.title('Распределение возраста отточных клиентов')
plt.xlabel('Возраст')
plt.ylabel('Частота встречаемости')
plt.show()
```

```python
bank_scrooge.query('city == "Ярославль" and gender == "М" and churn == 0')['age'].hist()

plt.title('Распределение возраста оставшихся клиентов')
plt.xlabel('Возраст')
plt.ylabel('Частота встречаемости')
plt.show()
```

В Ярославле возраст высокоотточных клиентов-мужчин составляет 30-40 лет и точно такой же возраст чаще встречается среди оставшихся клиентов.

```python
bank_scrooge.query('city == "Ростов" and gender == "М" and churn == 1')['age'].hist()

plt.title('Распределение возраста отточных клиентов')
plt.xlabel('Возраст')
plt.ylabel('Частота встречаемости')
plt.show()
```

```python
bank_scrooge.query('city == "Ростов" and gender == "М" and churn == 0')['age'].hist()

plt.title('Распределение возраста оставшихся клиентов')
plt.xlabel('Возраст')
plt.ylabel('Частота встречаемости')
plt.show()
```

В Ростове возраст высокоотточных клиентов-мужчин составляет 32-37 лет, среди оставшихся клиентов наиболее распространенный возраст -- 34-44 года.

```python
# Построим 2 датафрейма отточных клиентов:
# a - для мужчин из Ярославля 30-40 лет
# b - для мужчин из Ростова 32-37 лет
churn_clients_a = bank_scrooge.query('city == "Ярославль" and gender == "М" and churn == 1 and age >= 30 and age <= 40')
churn_clients_b = bank_scrooge.query('city == "Ростов" and gender == "М" and churn == 1 and age >= 32 and age <= 37')
```

```python
# Построим 2 датафрейма оставшихся клиентов:
# c - для мужчин из Ярославля 30-40 лет
# d - для мужчин из Ростова 34-44 лет
notchurn_clients_c = bank_scrooge.query('city == "Ярославль" and gender == "М" and churn == 0 and age >= 30 and age <= 40')
notchurn_clients_d = bank_scrooge.query('city == "Ростов" and gender == "М" and churn == 0 and age >= 34 and age <= 44')
```

```python
# Построим распределение количества клиентов по продуктам(Ярославль)
prod_a = churn_clients_a.groupby('products', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'}).sort_values(by='clients_count', ascending=False)

display(prod_a)

sns.barplot(x='products', y='clients_count', data=prod_a)

plt.title('Количество клиентов по продуктам')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество клиентов')

plt.show()
```

```python
# Построим распределение количества клиентов по продуктам(Ростов)
prod_b = churn_clients_b.groupby('products', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'}).sort_values(by='clients_count', ascending=False)

display(prod_b)

sns.barplot(x='products', y='clients_count', data=prod_b)

plt.title('Количество клиентов по продуктам')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество клиентов')

plt.show()
```

И в Ярославле, и в Ростове отточные клиенты-мужчины в большинстве случаев имели 2 банковских продукта.

```python
# Построим распределение количества клиентов по продуктам(Ярославль)
prod_c = notchurn_clients_c.groupby('products', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'}).sort_values(by='clients_count', ascending=False)

display(prod_c)

sns.barplot(x='products', y='clients_count', data=prod_c)

plt.title('Количество клиентов по продуктам')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество клиентов')

plt.show()
```

```python
# Построим распределение количества клиентов по продуктам(Ростов)
prod_d = notchurn_clients_d.groupby('products', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'}).sort_values(by='clients_count', ascending=False)

display(prod_d)

sns.barplot(x='products', y='clients_count', data=prod_d)

plt.title('Количество клиентов по продуктам')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество клиентов')

plt.show()
```

Оставшиеся клиенты в Ярославле чаще всего имеют 1 продукт, а в Ростове(как и среди отточных клиентов) -- 2 продукта.

```python
# Вычислим количество клиентов по наличию кредитной карты(Ярославль)
credit_card_a = churn_clients_a.groupby('credit_card', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'})

display(credit_card_a)

sns.barplot(x='credit_card', y='clients_count', data=credit_card_a)

plt.title('Количество клиентов по наличию кредитной карты')
plt.xlabel('Наличие кредитной карты(1-есть, 0-нет)')
plt.ylabel('Количество клиентов')

plt.show()
```

```python
# Вычислим количество клиентов по наличию кредитной карты(Ростов)
credit_card_b = churn_clients_b.groupby('credit_card', as_index=False).agg({'user_id': 'count'})\
.rename(columns={'user_id': 'clients_count'})

display(credit_card_b)

sns.barplot(x='credit_card', y='clients_count', data=credit_card_b)

plt.title('Количество клиентов по наличию кредитной карты')
plt.xlabel('Наличие кредитной карты(1-есть, 0-нет)')
plt.ylabel('Количество клиентов')

plt.show()
```

В обоих городах отточные клиенты-мужчины, более чем в 60% случаев, имели кредитную карту.

## 4. Проверка статистических гипотез

### 4.1. Составить 2 нулевые гипотезы

Гипотеза различия дохода между теми клиентами, которые ушли, и теми, которые остались.

H_0: Доходы отточных и неотточных клиентов не равны.

H_1: Доходы отточных и неотточных клиентов равны.

Гипотеза различия приблизительной оценки собственности между теми клиентами, которые ушли, и теми, которые остались.

H_0: Приблизительные оценки собственности отточных и неотточных клиентов не равны.

H_1: Приблизительные оценки собственности отточных и неотточных клиентов равны.

### 4.2. Провести проверку гипотез

```python
# Проверка гипотезы о доходах
churn_salary = bank_scrooge.query('churn == 1')['est_salary'] 
notchurn_salary = bank_scrooge.query('churn == 0')['est_salary']

alpha = .05

results = st.ttest_ind(
    churn_salary, 
    notchurn_salary,
    equal_var=False
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')   
```

В результате проверки гипотезы выяснено, что доходы отточных и неотточных клиентов различаются.

```python
# Проверка гипотезы об оценках собственности
churn_equity = bank_scrooge.query('churn == 1')['equity']
notchurn_equity = bank_scrooge.query('churn == 0')['equity']

alpha = .05

results = st.ttest_ind(
    churn_equity, 
    notchurn_equity,
    equal_var=False
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

В результате проверки гипотезы выяснено, что приблизительные оценки собственности отточных и неотточных клиентов одинаковые.
